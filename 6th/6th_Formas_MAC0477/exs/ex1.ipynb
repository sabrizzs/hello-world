Exercício 1: Visualização de imagens

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

# Função para ler os dados do CSV e carregar as imagens
def carregar_base(csv_path, img_dir):
    # Ler o CSV
    dados = pd.read_csv(csv_path)
    
    # Lista para armazenar os dados das imagens
    imagens = []
    
    # Iterar sobre as linhas do CSV
    for index, row in dados.iterrows():
        # Caminho da imagem
        img_path = os.path.join(img_dir, row['Nome da imagem'] + '.png')
        
        # Verificar se a imagem existe
        if os.path.exists(img_path):
            # Carregar a imagem
            imagem = plt.imread(img_path)
            
            # Adicionar os dados da imagem
            imagens.append({'imagem': imagem, 'dados': row})
    
    return imagens

# Função para exibir uma amostra das imagens de uma classe específica
def exibir_amostra_classe(imagens, classe, quantidade=5):
    # Filtrar as imagens da classe específica
    imagens_classe = [img['imagem'] for img in imagens if classe in img['dados'].values]
    
    # Exibir as imagens
    plt.figure(figsize=(15, 5))
    for i in range(min(quantidade, len(imagens_classe))):
        plt.subplot(1, quantidade, i+1)
        plt.imshow(imagens_classe[i])
        plt.title(f'Amostra {i+1}')
        plt.axis('off')
    plt.show()

# Função para exibir todas as classes com uma imagem exemplo
def exibir_classes(imagens):
    # Extrair todas as classes presentes nas imagens
    classes = set()
    for img in imagens:
        classes.update(img['dados'][3:].dropna())
    
    # Exibir as imagens exemplo de cada classe
    plt.figure(figsize=(15, 5))
    for i, classe in enumerate(classes):
        plt.subplot(1, len(classes), i+1)
        for img in imagens:
            if classe in img['dados'].values:
                plt.imshow(img['imagem'])
                plt.title(classe)
                plt.axis('off')
                break
    plt.show()

# Função para exibir a tabela sumária dos dados
def tabela_sumaria(imagens):
    # Contagem de classes e imagens
    classes = set()
    total_imagens = len(imagens)
    for img in imagens:
        classes.update(img['dados'][3:].dropna())
    total_classes = len(classes)
    
    # Tamanho total da base em bytes
    tamanho_total_bytes = sum(img['imagem'].size * img['imagem'].itemsize for img in imagens)
    
    # Resolução das imagens
    resolucao = np.mean([img['imagem'].shape[:2] for img in imagens], axis=0)
    
    # Exibir a tabela sumária
    print("=" * 35)
    print("Tabela Global Sumária:")
    print("● Nome das classes:", ', '.join(classes))
    print("● Número de classes:", total_classes)
    print("● Número de imagens:", total_imagens)
    print("● Tamanho da base (bytes):", tamanho_total_bytes / 1e6, "MB")
    print("● Resolução das imagens:", f"{int(resolucao[0])} linhas por {int(resolucao[1])} colunas")
    print("=" * 35)

# Função para exibir a tabela detalhada por classe
def tabela_detalhada(imagens):
    # Extrair informações detalhadas por classe
    detalhes_classe = {}
    for img in imagens:
        for classe in img['dados'][3:].dropna():
            if classe not in detalhes_classe:
                detalhes_classe[classe] = {
                    'Número de objetos': 0,
                    'Variações de fundo': set(),
                    'Variações de iluminação': set(),
                    'Número de repetições': 0,
                    'Total de amostras': 0
                }
            detalhes_classe[classe]['Número de objetos'] = max(detalhes_classe[classe]['Número de objetos'], sum([1 for obj in img['dados'][13:].dropna()]))
            detalhes_classe[classe]['Variações de fundo'].add(img['dados'][1])
            detalhes_classe[classe]['Variações de iluminação'].add(img['dados'][2])
            detalhes_classe[classe]['Número de repetições'] += 1
            detalhes_classe[classe]['Total de amostras'] += 1
    
    # Exibir a tabela detalhada
    print("=" * 110)
    print("{:<20} {:<15} {:<30} {:<25} {:<20} {:<15}".format('Nome da classe', 'Número de objetos', 'Variações de fundo', 'Variações de iluminação', 'Número de repetições', 'Total de amostras'))
    print("=" * 110)
    for classe, detalhes in detalhes_classe.items():
        print("{:<20} {:<15} {:<30} {:<25} {:<20} {:<15}".format(classe, detalhes['Número de objetos'], ', '.join(detalhes['Variações de fundo']), ', '.join(detalhes['Variações de iluminação']), detalhes['Número de repetições'], detalhes['Total de amostras']))
    print("=" * 110)

# Exemplo de uso das funções
imagens = carregar_base('dados.csv', 'caminho/para/as/imagens')
tabela_sumaria(imagens)
tabela_detalhada(imagens)
exibir_amostra_classe(imagens, 'caneta')
exibir_classes(imagens)
```
