#### Log Transformations (pág. 124)

- A transformação logarítmica comprime a faixa de valores de intensidade mais altos da imagem, enquanto expande a faixa de valores mais baixos.

- Ela é representada pela equação **s=c*log(1+r)**, onde s é a intensidade de saída, r é a intensidade de entrada e c é uma constante de ajuste.

- É útil para realçar detalhes em áreas de baixa intensidade em uma imagem, melhorando o contraste global e tornando os detalhes em áreas escuras mais visíveis.

- **Deixa áreas escuras mais visíveis.**



#### Power-Law Transformations (Gamma) (pág. 125)

- A transformação power-law, também conhecida como transformação gamma, é uma operação não linear que mapeia a entrada de intensidades para a saída de intensidades usando uma função exponencial.

- Ela é representada pela equação **s=c×r^ γ**, onde s é a intensidade de saída, r é a intensidade de entrada, c é uma constante de ajuste e γ é o parâmetro gamma.

- Pode ser usada para ajustar o contraste de uma imagem, expandindo ou comprimindo diferentes faixas de intensidades, dependendo do valor de γ. Valores de γ<1 expandem a faixa de intensidades mais escuras, enquanto valores de γ>1 comprimem essa faixa.





#### Linear Transformation

- A transformação linear é uma operação simples que realiza uma **mudança uniforme na escala de intensidades de uma imagem**.

- Ela é representada pela equação **s=cr+b**, onde s é a intensidade de saída, r é a intensidade de entrada, c é o coeficiente de ganho e b é o termo de viés.

    - coeficiente de ganho: taxa na qual as intensidades de entrada são mapeadas para as intensidades de saída
    - viés: termo de viés positivo pode ser usado para aumentar o brilho geral da imagem, enquanto um termo de viés negativo pode ser usado para diminuir o brilho

- Essa transformação não altera a forma geral do histograma da imagem e pode ser usada para ajustar o brilho e o contraste de maneira uniforme.

#### Bit-Plane Slicing (pág. 131)

- A técnica de "bit-plane slicing" é usada no processamento de imagens para decompor uma imagem em seus planos de bits individuais.

- Cada plano de bit representa um nível diferente de detalhe da imagem, indo do bit mais significativo (plano de alta ordem) ao bit menos significativo (plano de baixa ordem).

- O plano 1 é o do bit menos significativo.

- **Decomposição**: Uma imagem é decomposta em seus planos de bits, onde cada plano representa uma posição de bit específica dos valores dos pixels. Por exemplo, em uma imagem em tons de cinza de 8 bits, existem oito planos de bits, sendo que cada plano representa um dos oito bits nos valores dos pixels.



- **Análise da Importância da Imagem**: O "bit-plane slicing" ajuda na análise da importância relativa de cada bit na imagem. Essa análise ajuda a determinar a adequação do número de bits usados para representar a imagem. Os planos de alta ordem muitas vezes contêm dados visualmente significativos, enquanto os planos de baixa ordem contribuem para detalhes de intensidade mais sutis.



- **Compressão de Imagem**: A decomposição em planos de bits é útil para técnicas de compressão de imagens. Em vez de armazenar a imagem inteira, menos planos de bits podem ser armazenados e usados para reconstruir a imagem. Selecionando planos de bits específicos, a imagem pode ser reconstruída com requisitos de armazenamento reduzidos. Por exemplo, armazenar apenas os quatro planos de bits de maior ordem em vez da imagem original pode economizar 50% do espaço de armazenamento, mantendo um detalhamento de imagem aceitável.

- **Reconstrução**: Para reconstruir a imagem, os valores dos pixels são obtidos combinando os valores dos planos de bits correspondentes. Cada plano de bit é multiplicado por uma constante correspondente e os resultados são somados para obter a imagem em escala de cinza reconstruída.

Pseudocódigo do Bit-Plane Slicing:

função BitPlaneSlicing(imagem):
    array_de_planos[]

    para cada num_plano de 0 a 7:
        plano <- cria uma matriz com a mesma forma que a imagem

        para cada pixel [i][j] da imagem:
            valor_do_pixel <- imagem[i][j]
            bits <- converta valor_do_pixel para binário

            para cada bit de 0 a 7:
                se bit == num_plano então:
                    bits[bit] = bits[bit]
                senão:
                    bits[bit] = 0

            novo_pixel <- converta o valor de bits em decimal
            plano[i][j] <- novo_pixel

        adicione plano ao array_de_planos

    retorna array_de_planos

#### Histogram matching (specification) (pág 140)

- Enquanto a equalização de histograma busca produzir uma imagem de saída com um histograma uniforme, o Histogram Matching permite que nós determinemos a forma do histograma que desejamos que a imagem processada tenha.

- r e z representam os níveis de intensidade das imagens de entrada e saída, respectivamente.

- Podemos estimar pr(r) a partir da imagem de entrada fornecida, e pz(z) é a PDF especificada que desejamos que a imagem de saída tenha.

- definimos uma função G em relação a z com a propriedade G(z)=s. Isso nos leva à condição z=G(s).

- O processo de Especificação de Histograma implica em quatro etapas:
    - Obter pr(r) da imagem de entrada e utilizar na função de transformação T(r).
    - Utilizar a PDF especificada pz(z) para obter a função G(z).
    - Calcular a transformação inversa z=G^−1(s), mapeando os valores de s para z.
    - Obter a imagem de saída, primeiro equalizando a imagem de entrada e, em seguida, aplicando a transformação inversa para cada pixel.

#### Lowpass gaussian filter kernels

- O kernel gaussiano é uma técnica de filtragem utilizada em processamento de imagens para suavizar ou "borrar" uma imagem. É amplamente utilizado em aplicações onde é desejável reduzir o ruído e as bordas indesejadas em uma imagem.

- A principal característica do kernel gaussiano é sua distribuição de pesos, que segue uma função gaussiana.

- Esta função tem uma forma de sino e é caracterizada por sua suavidade e simetria circular, o que significa que sua resposta é independente da orientação.



- Uma das vantagens do kernel gaussiano é sua capacidade de preservar melhor as bordas da imagem em comparação com filtros box filter kernels.

-  Isso ocorre porque os pesos são distribuídos de acordo com a distância dos pixels em relação ao pixel central, de forma que os pixels mais próximos têm mais peso do que os pixels mais distantes. Isso faz com que as bordas da imagem sejam menos afetadas pelo processo de suavização.

- O kernel gaussiano é separável, o que significa que pode ser decomposto em duas operações de convolução unidimensionais. 

- Os kernels gaussianos possuem propriedades matemáticas úteis, como a propriedade de que a convolução de dois kernels gaussianos resulta em outro kernel gaussiano.

#### Order-statistic (nonlinear) filters

- Filtros Order-statistic (nonlinear) são filtros espaciais não lineares cuja resposta é baseada na ordenação (classificação) dos pixels contidos na região abrangida pelo filtro. 

- O suavização é alcançada substituindo o valor do pixel central pelo valor determinado pelo resultado da classificação. 

- O filtro mais conhecido nesta categoria é o filtro de mediana, que, como o nome indica, substitui o valor do pixel central pela mediana dos valores de intensidade no entorno desse pixel (o valor do pixel central é incluído no cálculo da mediana). 

- Os filtros de mediana fornecem excelentes capacidades de redução de ruído para certos tipos de ruído aleatório (ruído sal e pimenta), com consideravelmente menos borramento do que os filtros de suavização linear de tamanho semelhante. 

- A principal função dos filtros de mediana é fazer com que os pontos se pareçam mais com seus vizinhos.

#### Convolution

- O Teorema da Convolução é um resultado fundamental na teoria de Fourier que estabelece uma relação direta entre a convolução de duas funções no domínio do tempo (ou espacial) e o produto de suas transformadas de Fourier no domínio da frequência.

**Transformada de Fourier**

- A Transformada de Fourier transforma uma função do domínio do tempo (ou espacial) para o domínio da frequência. 

- Ela decompõe a função original em uma soma (ou integral) de funções senoidais, cada uma com uma determinada frequência, amplitude e fase.

**Convolução**

- A convolução de duas funções f(t) e g(t) é uma operação que combina as duas funções de uma maneira que captura como a forma de uma função é modificada pela outra.

- A convolução (f ★ g)(t) é definida como:



**Intuição por Trás do Teorema da Convolução**

- **Decomposição em Frequências:** Quando aplicamos a Transformada de Fourier a uma função, estamos essencialmente decompondo essa função em componentes senoidais de diferentes frequências. Cada função pode ser vista como uma soma (ou integral) de várias ondas senoidais.

- **Efeito da Convolução no Domínio do Tempo:** A convolução de duas funções no domínio do tempo combina essas funções de maneira que a saída em um dado momento t é uma espécie de média ponderada das funções f e g, onde uma delas está invertida e transladada. Essa operação é complexa no domínio do tempo, mas simplifica-se consideravelmente no domínio da frequência.

- **Produto no Domínio da Frequência:** No domínio da frequência, cada componente senoidal de f(t) e g(t) é multiplicado diretamente. Intuitivamente, a convolução no tempo combina todas as frequências das duas funções, o que se traduz em uma multiplicação simples de suas representações em frequência.

- **Simplificação da Convolução:** A convolução no tempo é uma operação integral que pode ser complicada de calcular diretamente. No entanto, a Transformada de Fourier converte a convolução em uma multiplicação de funções no domínio da frequência, que é uma operação mais simples e direta.

**Formalização do Teorema**

Matematicamente, o Teorema da Convolução afirma que se f(t) tem uma transformada de Fourier F(ω) e g(t) tem uma transformada de Fourier G(ω), então a convolução (f∗g)(t) tem uma transformada de Fourier dada pelo produto das transformadas de Fourier de f e g:



**Razão da Correspondência**

Essa correspondência entre convolução no tempo e produto no domínio da frequência pode ser entendida considerando a propriedade de linearidade e a natureza das funções senoidais. Cada senoide, quando convolvida com outra, resulta em uma modulação que pode ser descrita simplesmente multiplicando suas amplitudes e ajustando as frequências e fases. Assim, a convolução de duas funções no tempo, que mistura e sobrepõe suas informações, se reflete como uma simples multiplicação de suas componentes em frequência.

#### Teorema da Amostragem de Nyquist

Para digitalizar um sinal de áudio sem perda de informações, devemos seguir o Teorema da Amostragem de Nyquist, que afirma que a frequência de amostragem fs deve ser pelo menos duas vezes a maior frequência presente no sinal. Esta condição é necessária para evitar a aliasing (distorção) e garantir que o sinal original possa ser completamente recuperado da versão amostrada.

**Frequência mínima de amostragem**

Dado que o sinal de áudio possui frequências entre 20 Hz e 9 KHz, a maior frequência presente no sinal é 9 KHz. Assim, a frequência de amostragem mínima necessária é: fs ≥ 2×9 KHz = 18 KHz

Portanto, a frequência de amostragem mínima necessária é 18 KHz.

**Consequências da subamostragem em um sinal de imagem**

Subamostragem é o processo de amostrar um sinal a uma frequência menor do que a frequência de Nyquist (que é duas vezes a maior frequência presente no sinal). 

Quando isso ocorre em um sinal de imagem, várias consequências indesejadas podem surgir, principalmente devido ao fenômeno de aliasing. Vamos explorar essas consequências em detalhes:

Aliasing:

- Definição: Aliasing ocorre quando altas frequências no sinal original se confundem com frequências mais baixas no sinal amostrado. Isso causa a introdução de artefatos que não estavam presentes no sinal original.

- Aparência: Em imagens, isso pode se manifestar como padrões de moiré, serrilhamento em linhas diagonais, ou falsas texturas que distorcem a imagem.

Perda de Detalhe:

- Definição: A subamostragem pode resultar na perda de detalhes finos e texturas da imagem, pois a resolução amostral é insuficiente para capturar todas as variações presentes.

- Aparência: Detalhes pequenos, como cabelos, linhas finas ou texturas sutis, podem se tornar borrados ou desaparecer completamente.
