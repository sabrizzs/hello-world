### a - Mostre a imagem abaixo nos 8 planos referentes a cada um dos 8 bits em cada pixel.

```python
def int2bitarray(img):
    arr = []
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            arr.append(np.binary_repr(img[i][j], width=8))
    return arr
```
